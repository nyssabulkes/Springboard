## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# Import pandas
import pandas as pd

# Import Counter
from collections import Counter

# Import json
import json

# Import json_normalize
from pandas.io.json import json_normalize
```

```python
# Load in 'data/world_bank_projects.json' as Pandas dataframe
json_df = pd.read_json('data/world_bank_projects.json')
```

<b>TASK #1: Find the 10 countries with most projects</b>

```python
# Subset json_df: json_countries
json_countries = list(json_df['countryname'])

# Create a Counter object for the number of times each country appears: json_countries_counted
json_countries_counted = Counter(json_countries)

# Print list of 10 countries with the most projects
json_countries_counted.most_common(10)
```

<b>TASK #2: Find the top 10 major project themes</b>

```python
# Load json as string
json_data = json.load((open('data/world_bank_projects.json')))
    
# Normalize json_df column 'theme_namecode': json_normal
json_normal = json_normalize(json_data, 'mjtheme_namecode', ['id'])

# Group by themes: json_grouped
json_grouped = json_normal.groupby(['id']).first()

# Find top 10 most common themes
json_sorted = json_grouped.name.value_counts().head(10)
json_sorted

```

<b>TASK #3: Fill in the missing names</b>

```python
# Create new dataframe containing the theme names, country names, and ID: df1
df1 = json_normalize(json_data, 'mjtheme_namecode', ['id', 'countryname'])

# Create a second dataframe where there are no duplicates: df2
df2 = json_normalize(json_data, 'mjtheme_namecode').drop_duplicates()

# Create a third dataframe, which is df2 without empty strings in the name column: df3
df3 = df2[df2.name!='']

# Merge the original, df1, with the new df3: df4
df4 = df1.merge(df3, on='code')

# Create dataframe, with superfluous name column removed, reorganized for readability: df5
df5 = df4[['code', 'countryname', 'name_y', 'id']]
df5

```
